```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
import os

pytesseract.pytesseract.tesseract_cmd = r'/usr/local/bin/tesseract'

#function to extract text from PDF
def extract_text_from_pdf(pdf_path, output_txt_path):
    # Convert PDF to images
    images = convert_from_path(pdf_path, dpi=300)  # dpi set for better quality images

    extracted_text = ""

    for page_number, image in enumerate(images):
        print(f"Processing page {page_number + 1}...")

        text = pytesseract.image_to_string(image, lang='guj')  
        extracted_text += text + "\n"


    with open(output_txt_path, "w", encoding="utf-8") as f:
        f.write(extracted_text)

    print(f"Text extraction complete! Output saved to {output_txt_path}")
    return extracted_text


pdf_path = "Jamin Learning Doc.pdf"  
output_txt_path = "output_text.txt"    

extracted_text = extract_text_from_pdf(pdf_path, output_txt_path)
```

```python
import json

# Function to convert extracted text to JSON format
def convert_text_to_json(extracted_text, output_json_path):
    
    pages = extracted_text.split('\n\n')  

    
    data = {}
    for i, page_text in enumerate(pages, 1):
        data[f"page_{i}"] = page_text.strip()

    
    with open(output_json_path, "w", encoding="utf-8") as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

    print(f"Text successfully converted to JSON! Saved to {output_json_path}")
    return data


output_json_path = "output_text.json"  
convert_text_to_json(extracted_text, output_json_path)
```

```python
import json
import re


def load_json(json_path):
    with open(json_path, "r", encoding="utf-8") as json_file:
        return json.load(json_file)

# Search function to find relevant page(s) based on a query
def search_answer(query, json_data):
    # Normalize the query to lowercase
    query = query.lower()
    answers = []
    
    
    for page, text in json_data.items():
        if query in text.lower():
            answers.append((page, text))
    
    return answers

# Function to run the chatbot
def run_chatbot(json_data):
    while True:
        query = input("Ask a question (type 'exit' to quit): ")
        
        if query.lower() == "exit":
            print("Exiting chatbot...")
            break
        
        answers = search_answer(query, json_data)
        
        if answers:
            print(f"\nFound answers in the following pages:\n")
            for page, text in answers:
                print(f"--- {page} ---\n{text[:500]}...")  
        else:
            print("Sorry, I couldn't find an answer to that question.")


json_path = "output_text.json"  
json_data = load_json(json_path)  
run_chatbot(json_data)  
```
